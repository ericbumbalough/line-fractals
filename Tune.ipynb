```julia
using Interact, Reactive, Winston
include("LineFracGen.jl");
```

```julia
function tune3(points, n, span = .1, step = .005)
    #creates interactive sliders to change locations of 'points' [x1 y1;x2 y2...]
    #in a fractal with 'n' iterations. Slider has 'range' and 'step'.
    
    x2range = points[2,1]-span/2:step:points[2,1]+span/2
    y2range = points[2,2]-span/2:step:points[2,2]+span/2
    
    @manipulate for x2=x2range, y2=y2range
        data = frac([0 0; x2 y2; 1 0],n)
        p=plot(data[:,1],data[:,2])
        setattr(p.x1, draw_nothing = true)
        setattr(p.x2, draw_nothing = true)
        setattr(p.y1, draw_nothing = true)
        setattr(p.y2, draw_nothing = true)
        p
    end
end
```

```julia
function tune4(points, n, span = .1, step = .005)
    #creates interactive sliders to change locations of 'points' [x1 y1;x2 y2...]
    #in a fractal with 'n' iterations. Slider has 'range' and 'step'.
    
    x2range = points[2,1]-span/2:step:points[2,1]+span/2
    y2range = points[2,2]-span/2:step:points[2,2]+span/2
    x3range = points[3,1]-span/2:step:points[3,1]+span/2
    y3range = points[3,2]-span/2:step:points[3,2]+span/2
    
    @manipulate for x2=x2range, y2=y2range, x3=x3range, y3=y3range    
        data = frac([0 0; x2 y2; x3 y3; 1 0],n)
        p=plot(data[:,1],data[:,2])
        setattr(p.x1, draw_nothing = true)
        setattr(p.x2, draw_nothing = true)
        setattr(p.y1, draw_nothing = true)
        setattr(p.y2, draw_nothing = true)
        p
    end
end
```

```julia
function tuneclown(points, n, span = .1, step = .005)
    #creates interactive sliders to change locations of 'points' [x1 y1;x2 y2...]
    #in a fractal with 'n' iterations. Slider has 'range' and 'step'.
    
    x2range = points[2,1]-span/2:step:points[2,1]+span/2
    y2range = points[2,2]-span/2:step:points[2,2]+span/2
    
    @manipulate for x2=x2range, y2=y2range
        data = frac([0 0; x2 y2; 1-x2 y2; 1 0],n)
        p=plot(data[:,1],data[:,2])
        setattr(p.x1, draw_nothing = true)
        setattr(p.x2, draw_nothing = true)
        setattr(p.y1, draw_nothing = true)
        setattr(p.y2, draw_nothing = true)
        p
    end
end
```
