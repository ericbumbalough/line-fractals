```julia
using Images, ImageMagick
```

```julia
goodones = Float64[] #densities of choosen fractals
allones = Float64[] #densities of all fractals

for n=1:3171 #loop through all possible images
    path = pwd()*"\\Good Ones\\n=4\\"*string(n)*".png"
    
    if isfile(path) #file exists
        push!(goodones,mean(raw(load(path)))) #save average
    end
    
    path ="C:\\Plots\\"*string(n)*".png"
    if isfile(path) #file exists
        push!(allones,mean(raw(load(path)))) #save average
    end
end
        
```
